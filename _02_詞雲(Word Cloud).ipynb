```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 
word_freq = {'Python': 50, '':40, '': 30, '': 20, '': 5}

# 
wc = WordCloud(
    font_path='./fonts/NotoSansTC-Regular.otf',
    background_color='white',
    width=1024,
    height=768,
)

wc.generate_from_frequencies(word_freq) #  dict

# 
plt.figure(figsize=(10.24, 7.68), dpi=100) #=dpi*size
plt.imshow(wc, interpolation='bilinear')
plt.axis('off') # x, y
plt.show()
```

```python
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image # 
import matplotlib.pyplot as plt
import numpy as np
import jieba

# 
image = Image.open('./images/doraemon_01.png')
mask = np.array(image) # Np

text = '''
 
































































'''

seg_list = jieba.lcut(text)

seg_str = ' '.join(seg_list)

wc = WordCloud(
    font_path='./fonts/NotoSansTC-Regular.otf',
    background_color='white',
    mask=mask,
    width=820,
    height=1349,
    random_state=1234,
)

wc.generate(seg_str) # 

image_colors = ImageColorGenerator(mask)

wc.recolor(color_func=image_colors)

plt.figure(figsize=(8.2, 13.49), dpi=100)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
